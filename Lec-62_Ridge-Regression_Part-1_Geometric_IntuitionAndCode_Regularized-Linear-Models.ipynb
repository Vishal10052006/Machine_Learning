```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes

data = load_diabetes()
```

```python
#.DESCR is an attribute (not a function).
'''This includes:
What the dataset is about
Number of samples
Number of features
Feature meanings
Target meaning
Sometimes references / notes'''

print(data.DESCR)
```

```python
'''.data -
    A 2D NumPy array

    Contains only input features

    Shape: (number_of_samples, number_of_features)'''

'''.target - 
    A 1D NumPy array

    Contains the output values

    What the model must predict'''

X = data.data
y = data.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 45)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error

print("r2 score = ", r2_score(y_test, y_pred))
print("RMSE = ",np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
from sklearn.linear_model import Ridge

r = Ridge(alpha =0.00001)
```

```python
r.fit(X_train, y_train)
```

```python
print(r.coef_)
print(r.intercept_)
```

```python
y_pred1 = r.predict(X_test)
```

```python
print("R2 score",r2_score(y_test,y_pred1))
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred1)))
```

```python
m = 100
x1 = 5 * np.random.rand(m, 1) - 2
x2 = 0.7 * x1 ** 2 - 2 * x1 + 3 + np.random.randn(m, 1)

plt.scatter(x1, x2)
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

def get_preds_ridge(x1, x2, alpha):
    model = Pipeline([
        ('poly_feats', PolynomialFeatures(degree=16)),
        ('ridge', Ridge(alpha=alpha))
    ])
    model.fit(x1, x2)
    return model.predict(x1)

alphas = [0, 10, 100]
cs = ['r', 'g', 'b']

plt.figure(figsize=(10, 6))
plt.plot(x1, x2, 'b+', label='Datapoints')

for alpha, c in zip(alphas, cs):
    preds = get_preds_ridge(x1, x2, alpha)
    # Plot
    plt.plot(sorted(x1[:, 0]), preds[np.argsort(x1[:, 0])], c, label='Alpha: {}'.format(alpha))

plt.legend()
plt.show()
```
