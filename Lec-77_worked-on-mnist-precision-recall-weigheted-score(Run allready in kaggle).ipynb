```python
df = pd.read_csv('/kaggle/input/datasets/wasiqaliyasir/digital-mnist-dataset/mnist.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=2)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
lor = LogisticRegression()
dt = DecisionTreeClassifier()
```

```python
lor.fit(X_train, y_train)
dt.fit(X_train, y_train)
```

```python
y_pred1 = lor.predict(X_test)
y_pred2 = dt.predict(X_test)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
precision_score(y_test, y_pred1, average='macro')
```

```python
recall_score(y_test, y_pred2, average='macro')
```

```python
f1_score(y_test, y_pred2, average='macro')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred1))
```
