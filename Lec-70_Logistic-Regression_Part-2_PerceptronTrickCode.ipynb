```python
from sklearn.datasets import make_classification
import numpy as np

X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0,
                           hypercube=False, n_classes=2, n_clusters_per_class=1, random_state=41, class_sep=10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=100)
```

```python
def perceptron(X, y):

    X = np.insert(X, 0, 1, axis=1)   # hum kyu kar rahe h 1, taki boundary can shift any where, esse model me flexibility aa jati h. origin ke along nhi ghumti h. 
    weights = np.ones(X.shape[1])    # yaha se hum ek line decide kar rhe taki, ek line drow ho sake, esi liye hum starting me 1 lete h, ab model 1 point lekar compare karegi line ke saath, is it <0 or >0.
    lr = 0.1

    for i in range(10000):

        j = np.random.randint(0, 100)
        y_hat = step(np.dot(X[j], weights))
        weights = weights + lr * (y[j] - y_hat) * X[j]

    return weights[0], weights[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_, coef_ = perceptron(X, y)
```

```python
print(intercept_)
print(coef_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
X_input = np.linspace(-3, 3, 100)
y_input = m*X_input
```

```python
plt.figure(figsize=(10, 6))
plt.plot(X_input, y_input, color='red', linewidth=3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=100)
plt.ylim(-3, 2)
```

# For Animation Code

```python
def perceptron(X, y):
    
    m = []
    b = []
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(200):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
        m.append(-(weights[1]/weights[2]))
        b.append(-(weights[0]/weights[2]))
        
    return m,b
```

```python
m,b = perceptron(X,y)
```

```python

from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
```

```python
fig, ax = plt.subplots(figsize=(9,5))

x_i = np.arange(-3, 3, 0.1)

line, = ax.plot(x_i, x_i*m[0] + b[0], 'r-', linewidth=2)

ax.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
plt.ylim(-3,3)
def update(i):
    line.set_ydata(x_i*m[i] + b[i])
    ax.set_xlabel(f"epoch {i+1}")
    return line,

anim = FuncAnimation(fig, update, frames=200, interval=100, blit=False)
plt.show()
```
